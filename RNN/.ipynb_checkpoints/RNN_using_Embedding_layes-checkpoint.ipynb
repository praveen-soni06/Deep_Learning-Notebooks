```python
from tensorflow.keras.datasets import imdb
from keras import Sequential
from tensorflow.keras.layers import Dense,SimpleRNN, Embedding, Flatten, Input
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import pad_sequences
```

```python
(X_train,y_train),(X_test,y_test) = imdb.load_data(num_words=10000)
```

```python
# trimming and adding pad

X_train = pad_sequences(X_train, padding='post', maxlen=100)
X_test = pad_sequences(X_test, padding='post', maxlen=100)
```

```python
model = Sequential()

model.add(SimpleRNN(32,input_shape=(100,1),return_sequences=False))
model.add(Dense(1,activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```

```python
model.predict(X_test)
```

# Using Embedding layer of keras

```python
docs = ['go india',
'india india',
'hip hip hurray',
'jeetega bhai jeetega india jeetega',
'bharat mata ki jai',
'kohli kohli',
'sachin sachin',
'dhoni dhoni',
'modi ji ki jai',
'inquilab zindabad']
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(docs)
```

```python
len(tokenizer.word_index)
```

```python
sequences = tokenizer.texts_to_sequences(docs)
sequences
```

```python
sequences = pad_sequences(sequences, padding='post')
sequences
```

```python
model = Sequential()
model.add(Input(shape=(5,)))
model.add(Embedding(18,output_dim=2, input_length=5))

model.summary()
```

```python
model.compile('adam', 'accuracy')

pred = model.predict(sequences)
print(pred)
```

```python
model = Sequential()
model.add(Input(shape=(100,)))
model.add(Embedding(10000,2))
model.add(SimpleRNN(32,return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(X_train, y_train,epochs=5, validation_data=(X_test,y_test))
```

```python
model.predict(X_test)
```
