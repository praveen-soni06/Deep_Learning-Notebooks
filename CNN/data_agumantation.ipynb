```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

```python
img = image.load_img(r"C:\Users\prabh\Downloads\ChatGPTImageMay10202601_27_01P.jpeg", target_size=(200,200))
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i = 0
for output in datagen.flow(input_batch, batch_size=1, save_to_dir=r"C:\Users\prabh\Downloads\aug_img"):
    i = i+1
    if i == 50:
        break
```
