```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# dl imports
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU,PReLU,ELU
```

```python
df = pd.read_csv('Churn_Modelling.csv')
print(df.shape)
print(df.info())
print(df.isnull().sum())
display(df.head(5))
```

```python
# seprate dependent and independent variables

X = df.iloc[:,3:13]
y = df.iloc[:,-1]
```

```python
# feature engineering

Geography = pd.get_dummies(X['Geography'], drop_first=True, dtype=int)
Gender = pd.get_dummies(X['Gender'], drop_first=True, dtype=int)
```

```python
X.drop(['Geography', 'Gender'], axis=1, inplace=True)
```

```python
X = pd.concat([X, Geography,Gender], axis=1)
```

```python
# feature engineering

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)

sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

print(X_train)
print("\n\n",X_test)
print('\n\n', X_train.shape)
```

# Second part ANN

```python
# intialize the ANN
clf = Sequential()
```

```python
# initialize the input layers
clf.add(Dense(units=11, activation='relu'))

# adding the first hidden layers
clf.add(Dense(units=7, activation='relu'))

# adding the second hidden layers
clf.add(Dense(units=6, activation='relu'))

# adding the output layers
clf.add(Dense(units=1, activation='sigmoid'))
```

```python
clf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
## early stoping
import tensorflow as tf
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    min_delta=0.0001,
    patience=20,
    verbose=1,
    mode="auto",
    baseline=None,
    restore_best_weights=False,
    start_from_epoch=0,
)
```

```python
model_history = clf.fit(X_train, y_train, validation_split=0.33, batch_size=10, epochs=1000, callbacks=early_stopping)
```

```python
model_history.history.keys()
```

```python
# summarize history for accuracy

plt.plot(model_history.history['accuracy' ])
plt.plot(model_history.history['val_accuracy' ])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# part 3 - making the predictionas and evaluating the model

# prediction the test set result
y_pred = clf.predict(X_test)
y_pred = (y_pred >= 0.5)
```

```python
# make the confusion matrix

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
# clculate the accureccy

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
